```python
produklist = []
baseurl = "https://www.dexagroup.com/id/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari.537.36"
}
```

```python
productlinks = []
for x in range(1, 24):
    r = requests.get(f"https://www.dexagroup.com/id/semua-produk-dexa-group/?sf_paged={x}")
    soup = BeautifulSoup(r.content, "lxml")
    productlist = soup.find_all("div", class_="et_pb_code_inner")

    for item in productlist:
        for link in item.find_all("a", href=True):
            productlinks.append(link["href"])
```

```python
productlinks = list(dict.fromkeys(productlinks))
for x in range (2, 24):
    productlinks.remove(f"https://www.dexagroup.com/id/semua-produk-dexa-group/?sf_paged={x}")
productlinks.remove("https://www.dexagroup.com/id/semua-produk-dexa-group/")
```

```python
for link in productlinks:
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content, "lxml")

    nama = soup.find("h1").text.strip()
    komposisi = soup.find("tbody").text.strip()
    #kandungan = komposisi.text.strip()
    #gram = komposisi.find_next_sibling("td").text.strip()
    indikasi = soup.find("div", class_="et_pb_column et_pb_column_4_4 et_pb_column_inner et_pb_column_inner_21_tb_body et-last-child").text.strip()
    produk = {
        "nama": nama,
        "komposisi": komposisi,
        "indikasi": indikasi
    }
    produklist.append(produk)
    print("Saving: ", produk["nama"])
```

```python
df = pd.DataFrame(produklist)
df.to_csv("uji4.csv", index = False)
```
